```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("dark")
```

```python
url='https://raw.githubusercontent.com/maruthireddy432/Alzheimer_classification/main/alzheimers_disease_data.csv'
raw_data = pd.read_csv(url)
df=raw_data.copy()
```

```python
df.drop(['PatientID','DoctorInCharge'],axis=1,inplace=True)
df
```

## Visualizations

```python
columns = df[['Age','BMI','AlcoholConsumption','PhysicalActivity','DietQuality','SleepQuality','SystolicBP','DiastolicBP','CholesterolTotal','CholesterolLDL','CholesterolHDL','CholesterolTriglycerides','MMSE','FunctionalAssessment','ADL']].columns
fig, axes = plt.subplots(len(columns), 2, figsize=(18, len(columns)*5))
for i, col in enumerate(columns):
    sns.boxplot(x=df[col], ax=axes[i,0], data=df)
    axes[i,0].set_title(f"Boxplot of {col}")
    sns.histplot(df[col], bins=30, kde=True, ax=axes[i,1])
    axes[i,1].set_title(f"Distribution of {col}")
plt.show()
```

```python
sns.countplot(x=df['Gender'],data=df,hue=df['Diagnosis'])
plt.title('Diagonsis by Gender')
plt.legend(title="Diagnosis", labels=["No Alzheimer's", "Alzheimer's"])
plt.xlabel("Gender 0 = Female, 1 = Male")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
corr_matrix = df.corr()
sns.heatmap(corr_matrix, cmap="coolwarm", annot=False, linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df["Age"], y=df["MMSE"], hue=df["Diagnosis"], alpha=0.7)
plt.title("Age vs. MMSE Score", fontsize=14)
plt.xlabel("Age", fontsize=12)
plt.ylabel("MMSE Score", fontsize=12)
# plt.legend(labels=["No Alzheimer's","Alzheimer's"])
plt.show()
```
